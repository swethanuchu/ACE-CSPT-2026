Calculator using functions

```python
def add(a,b):
  return a+b
def sub(a,b):
  return a-b
def mul(a,b):
  return a*b
def div(a,b):
  return a/b
a=int(input("enter a number"))
b=int(input("enter a number"))
print("1.add(a,b)")
print("2.sub(a,b)")
print("3.mul(a,b)")
print("4.div(a,b)")
choice=int(input("enter your choice"))
if choice==1:
      print("addition:",add(a,b))
elif choice==2:
      print("subtraction:",sub(a,b))
elif choice==3:
      print("multiplication:",mul(a,b))
elif choice==4:
      print("division:",div(a,b))
```

Python code to solve two linear equations

```python
a1 = float(input("Enter coefficient of x: "))
b1 = float(input("Enter coefficient of y: "))
c1 = float(input("Enter constant of equation 1: "))
a2 = float(input("Enter coefficient of x in equation 2: "))
b2 = float(input("Enter coefficient of y in equation 2: "))
c2 = float(input("Enter constant of equation 2: "))
D = a1 * b2 - a2 * b1
if D == 0:
    print("No unique solution exists.")
else:
    x = (c1 * b2 - c2 * b1) / D
    y = (a1 * c2 - a2 * c1) / D
    print("x =", x)
    print("y =", y)
```
